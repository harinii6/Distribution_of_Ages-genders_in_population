```python
import pandas as pd

# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv(url)

titanic_data.head()
```

```python
import matplotlib.pyplot as plt

# Plotting the bar chart for gender distribution
gender_counts = titanic_data['Sex'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution on the Titanic')
plt.show()
```

```python
# Plotting the histogram for age distribution
plt.figure(figsize=(10, 6))
plt.hist(titanic_data['Age'].dropna(), bins=15, edgecolor='black', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution on the Titanic')
plt.show()
```
